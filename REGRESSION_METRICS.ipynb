```python
import pandas as pd
import numpy as np
from google.colab import files
files.upload()
df = pd.read_csv("placement.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['package'],color = 'red')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('package',axis = 1),df['package'],test_size  = 0.3,random_state= True)
x_train
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(x_train,y_train)
```

```python
linear.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
y_test
```

```python
import matplotlib.pyplot as plt
plt.plot(x_train,linear.predict(x_train),color = 'blue')
plt.scatter(df['cgpa'],df['package'],color = 'red')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_pred = linear.predict(x_test)
```

```python
y_test.values
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
print("MsE",mean_squared_error(y_test,y_pred))
```

```python
import numpy as np
print("RMSE",np.sqrt(mean_absolute_error(y_test,y_pred)))
```

```python
print("R2",r2_score(y_test,y_pred))
```

```python
r2 = r2_score(y_test,y_pred)
Adjusted_R2 = 1 - ((1-r2)*(x_test.shape[0] - 1)) / (x_test.shape[0] - 1 - 1)
print(Adjusted_R2)
```
