```python
df = pd.read_csv("multiple_linear_regression_dataset.csv")
df.head()
df.drop(columns = ['Age'],inplace = True)
df
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.scatter_3d(x = df['Experience'] , y = df['EducationLevel'] , z = df['Salary'])
fig.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('Salary',axis = 1), df['Salary'],train_size = 0.8,random_state= True)
x_train
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
lr.predict(x_test)
```

```python
import numpy as np

x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)
xGrid, yGrid = np.meshgrid(y, x)
final = np.vstack((xGrid.ravel(), yGrid.ravel())).T
z_final = lr.predict(final).reshape(10, 10)
z = z_final
```

```python
fig = px.scatter_3d(x = df['Experience'] , y = df['EducationLevel'] , z = df['Salary'])

fig.add_trace(go.Surface(x = x, y = y, z =z ))

fig.show()
```
